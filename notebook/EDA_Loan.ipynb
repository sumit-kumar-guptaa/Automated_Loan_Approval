```python
features = ['fico', 'int.rate', 'dti', 'revol.util', 'inq.last.6mths', 'credit.policy']
target = 'not.fully.paid'
```

```python
num_features = ['fico', 'int.rate', 'dti', 'revol.util']
for feature in num_features:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[feature], kde=True)
    plt.title(f'Distribution of {feature}')
    plt.show()
```

```python
cat_features = ['inq.last.6mths', 'credit.policy']
for feature in cat_features:
    plt.figure(figsize=(8, 4))
    sns.countplot(x=df[feature])
    plt.title(f'Count of {feature}')
    plt.show()
```

```python
print(df[features + [target]].describe())
```

```python
for feature in num_features:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=target, y=feature, data=df)
    plt.title(f'{feature} vs. {target}')
    plt.show()
```

```python
for feature in cat_features:
    plt.figure(figsize=(8, 4))
    sns.barplot(x=feature, y=target, data=df, estimator=lambda x: sum(x) / len(x))
    plt.title(f'Default Rate by {feature}')
    plt.ylabel('Default Rate')
    plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df[features + [target]].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
print(df[features].isnull().sum())
```

```python
# Outlier detection (IQR method)
for feature in num_features:
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[feature] < Q1 - 1.5 * IQR) | (df[feature] > Q3 + 1.5 * IQR)][feature]
    print(f'Outliers in {feature}: {len(outliers)}')
```
